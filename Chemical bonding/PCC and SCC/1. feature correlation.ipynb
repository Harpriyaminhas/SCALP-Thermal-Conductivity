```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/dataset2')
df
```

```python
df.columns
```

```python
columns_to_drop = ['Composition']
df= df.drop(columns=columns_to_drop)
df
```

```python
df = df.dropna()
df
```

```python
# Convert 'kappa.1' to numeric values
df['kappa'] = pd.to_numeric(df['kappa'], errors='coerce')

df_filtered = df[df['kappa'] <= 2]
```

```python
df_filtered
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
corr = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(15, 13))
plt.rcParams['figure.dpi'] = 1200
plt.xticks(weight='bold', fontsize=10)
plt.yticks(weight='bold', fontsize=10)

# Create a heatmap with 'coolwarm' palette and annotations
sns.heatmap(corr,
            cmap='coolwarm', vmax=1.0, vmin=-1.0, linewidths=0.1, square=True)

# Title and adjustments
plt.title('Correlation Heatmap', fontsize=14, weight='bold')
plt.tight_layout()

# Show the plot
plt.show()
```
