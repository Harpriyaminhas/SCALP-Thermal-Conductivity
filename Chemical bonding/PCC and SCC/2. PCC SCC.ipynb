```python
import pandas as pd
import numpy as np
```

Input Dataset

```python
df0 = pd.read_excel('/content/dataset3.xlsx')
df0
```

```python
columns_to_drop = ['Composition']
x = df0.drop(columns=columns_to_drop)
x
```

```python
x = x[x['kappa'] <= 2]
x
```

#PCC

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = x.corr()

# Set up the matplotlib figure
plt.figure(figsize=(15, 13))
plt.rcParams['figure.dpi'] = 500
plt.xticks(weight='bold', fontsize=10)
plt.yticks(weight='bold', fontsize=10)

# Create a heatmap
sns.heatmap(corr,
            cmap= 'RdYlBu', vmax=1.0, vmin=-1.0, linewidths=0.2, linecolor='brown',
            annot=True, fmt=".2f", annot_kws={"size": 9, "weight": "bold"}, square=True)

# Show the p
plt.show()
```
