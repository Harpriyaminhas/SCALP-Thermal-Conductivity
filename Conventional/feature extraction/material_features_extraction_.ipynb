```python
from pymatgen.core import Composition, Structure
```

```python
from matminer.featurizers.conversions import StrToComposition

stc = StrToComposition()
df_Comp = stc.featurize_dataframe(df, "formula", pbar=False)
df_Comp
```

```python
from matminer.featurizers.composition import ElementProperty

ep_feat = ElementProperty.from_preset(preset_name="magpie")
df2 = ep_feat.featurize_dataframe(df_Comp, col_id="composition")  
df2.head()
```

```python
from matminer.featurizers.base import MultipleFeaturizer
from matminer.featurizers import composition as cf
```

```python
feature_calculators = MultipleFeaturizer([cf.Stoichiometry(), cf.ValenceOrbital(props=['avg']), cf.IonProperty(fast=True)])
feature_labels = feature_calculators.feature_labels()
df3 = feature_calculators.featurize_dataframe(df2, col_id='composition');
df3.head()
```

```python
from matminer.featurizers.conversions import CompositionToOxidComposition
from matminer.featurizers.composition import OxidationStates
```

```python
df4 = CompositionToOxidComposition().featurize_dataframe(df3, "composition", ignore_errors=True)
df4.head()
```

```python
os_feat = OxidationStates()
df5 = os_feat.featurize_dataframe(df4, "composition_oxid", ignore_errors=True)
df5.head()
```

```python
from matminer.featurizers.composition import BandCenter
bc_feat = BandCenter()
df6 = bc_feat.featurize_dataframe(df5,"composition_oxid", ignore_errors=True)
df6
```

```python
from matminer.featurizers.composition import IonProperty
from matminer.featurizers.composition import ElectronAffinity
from matminer.featurizers.composition import ElectronegativityDiff
```

```python
eneg_diff = ElectronegativityDiff()

df_with_eneg_diff = eneg_diff.featurize_dataframe(df6, "composition_oxid", 'ignore_errors=True')
```
