## XGBR

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from math import sqrt
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming x and y are your feature matrix and target variable, respectively
# Replace them with your actual data

# Initialize variables to store the best metrics and corresponding random state
best_random_state = None
min_mae_test = float('inf')
min_rmse_test = float('inf')
max_r2_test = float('-inf')
min_mae_train = float('inf')
min_rmse_train = float('inf')
max_r2_train = float('-inf')

# Loop through different random states
for random_state in range(0, 2000):  # You can adjust the range as needed
    # Split the data into training and testing sets
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=random_state)

    # Initialize the XGBoost Regressor model
    xg_reg = xgb.XGBRegressor()

    # Train the model
    xg_reg.fit(x_train, y_train)

    # Predictions
    xg_pred_test = xg_reg.predict(x_test)
    xg_pred_train = xg_reg.predict(x_train)

    # Calculate metrics for test set
    xg_rmse_test = sqrt(mean_squared_error(y_test, xg_pred_test))
    xg_mae_test = mean_absolute_error(y_test, xg_pred_test)
    xg_r2_test = r2_score(y_test, xg_pred_test)

    # Calculate metrics for train set
    xg_rmse_train = sqrt(mean_squared_error(y_train, xg_pred_train))
    xg_mae_train = mean_absolute_error(y_train, xg_pred_train)
    xg_r2_train = r2_score(y_train, xg_pred_train)

    # Update the best metrics and random state if needed for test set
    if xg_mae_test < min_mae_test:
        min_mae_test = xg_mae_test
        min_rmse_test = xg_rmse_test
        max_r2_test = xg_r2_test
        best_random_state = random_state

    # Update the best metrics if needed for train set
    if xg_mae_train < min_mae_train:
        min_mae_train = xg_mae_train
        min_rmse_train = xg_rmse_train
        max_r2_train = xg_r2_train

# Print the best metrics and corresponding random state for test set
print("Best Random State (Test Set):", best_random_state)
print("Min Test RMSE:", min_rmse_test)
print("Min Test MAE:", min_mae_test)
print("Max Test R2:", max_r2_test)

# Print metrics for train set with the best random state
print("\nTrain Error Reports with the Best Random State:")
print("Min Train RMSE:", min_rmse_train)
print("Min Train MAE:", min_mae_train)
print("Max Train R2:", max_r2_train)

# Train the model with the best random state for test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=best_random_state)
xg_reg = xgb.XGBRegressor()
xg_reg.fit(x_train, y_train)
xg_pred_test = xg_reg.predict(x_test)

# Plotting the graph
plt.figure(figsize=(5.5, 5.5))

# Plotting the grid
plt.grid(True, linestyle='--', alpha=0.5)

# Plotting the train set
sns.regplot(x=y_train, y=xg_reg.predict(x_train), color='mediumblue', scatter_kws={'s': 70},
            line_kws={'color': 'khaki'}, label='Train set')

# Plotting the test set above the train set
plt.scatter(y_test, xg_pred_test, color='deeppink', s=70, label='Test set')

plt.xlabel('Predicted Kappa', fontsize=12, weight='bold')
plt.ylabel('Experimental Kappa', fontsize=12, weight='bold')
plt.gca().set_aspect('equal', adjustable='box')
plt.xticks(weight='bold')
plt.yticks(weight='bold')
plt.legend(fontsize=16)
plt.show()
```
